**Objetivo:**

Comparar empiricamente a eficiência de dois métodos de ordenação e identificar o ponto de ruptura onde a teoria assintótica se torna visível na pratica


---


**Exercício:**

- Implemente o algoritmo Insertion Sort (o n^2)

- Utilize a função nativa do python sorted() que usa o TimSort (n log n) para fins de comparação

- Gere listas aleatórias de tamanhos n= [1000, 5000, 10000, 20000, 50000]

- Meça o tempo de execução de ambos os algoritmos para cada n, gerando a saída no terminal de cada um deles - utilize a biblioteca time

```python
import time
import random
```

```python
"""
Performance benchmark comparing Insertion Sort and Python's native TimSort.
"""

SCALE = [1000, 5000, 10000, 20000, 50000]

def insertion_sort(arr):
  """
  Sorts a list of integers using insertion sort algorithm.
  :param arr: list of integers to be sorted.
  """

  for i in range (1, len(arr)):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j = j - 1

    arr[j+1] = key
```

```python
def run_sorting_benchmark():
  """
  Runs the performance test for each scale defined.
  """

  for n in SCALE:
    # Generate random data
    base_list = [random.randint(0, n * 10) for _ in range(n)]

    # Measuring insertion sort with TimSort (n log n)
    tim_list = base_list.copy()
    start_tim_sort = time.time()
    tim_list.sort()
    end_tim_sort = time.time()
    time_tim_sort = end_tim_sort - start_tim_sort

    # Measuring insertion sort manually (O(n^2))
    insertion_list = base_list.copy()
    start_insertion_sort = time.time()
    insertion_sort(insertion_list)
    end_insertion_sort = time.time()
    time_insertion_sort = end_insertion_sort - start_insertion_sort

    # Output
    print(f"n = {n} - Tempo para TimSort: {time_tim_sort:.6f}s")
    print(f"n = {n} - Tempo para Insertion Sort manual: {time_insertion_sort:.6f}s")

    print(f"n = {n:<5} | TimSort: {time_tim_sort:>10.6f}s | Insertion: {time_insertion_sort:>10.6f}s")

    print(f"-"*40)
```

```python

if __name__ == "__main__":
    run_sorting_benchmark()
```
