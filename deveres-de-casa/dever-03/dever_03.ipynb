Implemente o código recursivo para verificar se um array é palíndromo.

Exemplos:

array1 = [0, 1, 2, 3, 2, 1, 0] -> É palíndromo

array2 = ["a", "b", "b", "a"] -> É palíndromo

array3 = ["a", "b", "c", "b", "a"] -> É palíndromo

array4 = ["a", "b", "c", "f", "b", "a"] -> Não é palíndromo

```python
def ehPalindromo(array):
  if len(array) == 0 or len(array) == 1: # array vazio ou array com um caractere
    return True
  elif array[0] != array[-1]:
    return False
  else:
    return ehPalindromo(array[1:-1])


array1 = [0, 1, 2, 3, 2, 1, 0]
print(ehPalindromo(array1))
array2 = ["a", "b", "b", "a"]
print(ehPalindromo(array2))
array3 = ["a", "b", "c", "b", "a"]
print(ehPalindromo(array3))
array4 = ["a", "b", "c", "f", "b", "a" ]
print(ehPalindromo(array4))
```
